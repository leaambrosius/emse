```python
data = data.drop(['Zeitstempel', 'Anfangsbuchstabe des Vornamens Ihrer Mutter','Anfangsbuchstabe des Mädchennamens Ihrer Mutter',"Dritter Buchstabe des Geburtsmonats Ihrer Mutter","Anfangsbuchstabe Ihres Geburtsortes ", "Anfangsbuchstabe des Vornamens Ihres Vaters"], axis=1)
data.head(3)
```

```python
data.drop(data[(data['Wie häufig machen Sie in der Woche Sport?'] == "Gar nicht")].index, inplace=True)
data
```

```python
data = data.replace("1 - trifft gar nicht zu", 1)
data = data.replace("2 - trifft nicht zu", 2)
data = data.replace("3- trifft teilweise zu", 3)
data = data.replace("4 - trifft zu", 4)
data = data.replace("5 - trifft voll zu", 5)
```

```python
# calculate amotivation
amotivationArray = [] 
i = 0
for i in range(107):
    amotivation = (data.iloc[i,9] +data.iloc[i,12] + data.iloc[i,21] +data.iloc[i,26]) /4
    amotivationArray.append(amotivation)
data["Amotivation"] = amotivationArray
```
